```python
import numpy as np 
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import joblib

from sklearn.model_selection import train_test_split
from sklearn.dummy import DummyClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix,classification_report
from sklearn.preprocessing import MinMaxScaler, StandardScaler, OneHotEncoder
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
```

```python
titanic=pd.read_csv("/kaggle/input/datasets/fatuzahra/titanic/titanic.csv")
titanic.head()
```

```python
titanic["Age"].fillna(
    titanic["Age"].mean(),inplace=True
)


titanic["Embarked"].fillna(
    titanic["Embarked"].mode()[0], inplace=True
)

titanic= titanic.drop(columns="Cabin")

titanic.isnull().sum()
```

```python
titanic=titanic[['Survived','Pclass','Age','Fare','Sex']]

titanic['Age'].fillna(titanic['Age'].median(),inplace=True)


titanic['Sex']=titanic['Sex'].map({'male':0,'female':1})
```

```python
X= titanic.drop('Survived',axis=1)
Y=titanic['Survived']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y,
    test_size=0.2,
    random_state=42
)
```

```python
param_grid={
    'n_estimators':[50,100],
    'max_depth':[None,5,10],
    'min_samples_split':[2,5]
}

rf=RandomForestClassifier(random_state=42)
```

```python
grid=GridSearchCV(
    rf,
    param_grid,
    cv=5,
    scoring="accuracy"
)
```

```python
grid.fit(X_train,Y_train)
```

```python
print('Best parameters:',grid.best_params_)
print('Best Accuracy:',grid.best_score_)
```

```python
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
kmeans=KMeans(n_clusters=5,random_state=42)
clusters=kmeans.fit_predict(X_scaled)
titanic['Clusters']=clusters
titanic.head()
```

```python
wcss=[]

for k in range(1,11):
    kmeans=KMeans(n_clusters=k,random_state=42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11),wcss,marker='o')
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.title("Elbow Method")
plt.show()
```

```python
score=silhouette_score(X_scaled,clusters)
print("Silhouette Score:",score)
```

```python
for k in range( 2, 8):
    kmeans = KMeans(n_clusters = k, random_state = 42)
    labels = kmeans.fit_predict(X_scaled)
    score = silhouette_score(X_scaled, labels)
    print(f"k = {k}, Score: {score}")
```

```python
pca=PCA(n_components=2)
X_pca=pca.fit_transform(X_scaled)

plt.scatter(X_pca[:,0],X_pca[:,1],c=clusters)
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```

```python
best_model = grid.best_estimator_
joblib.dump(best_model,"titanic_rf_model.pk1")
joblib.dump(scaler,"scaler.pk1")
```

```python
titanic=pd.read_csv("/kaggle/input/datasets/fatuzahra/titanic/titanic.csv")
```

```python
Y=titanic['Survived']
X=titanic[['Pclass','Sex','Age','Fare','Embarked']]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y,
    test_size=0.2,
    random_state=42
)
```

```python
numeric_features=['Age','Fare','Pclass']
categorical_features=['Sex','Embarked']
```

```python
numeric_transformer=Pipeline([
    ('impute',SimpleImputer(strategy='mean')),
    ('scaler',StandardScaler())
])
```

```python
categorical_tranformer=Pipeline([
    ('impute',SimpleImputer(strategy='most_frequent')),
    ('encoder',OneHotEncoder())
])
```

```python
preprocessor=ColumnTransformer([
    ('num',numeric_transformer,numeric_features),
    ('cat',categorical_tranformer,categorical_features)
])
```

```python
pipeline=Pipeline([
    ('preprocessor',preprocessor),
    ('model',RandomForestClassifier())
])
```

```python
pipeline.fit(X_train,Y_train)
```

```python
joblib.dump(pipeline,"titanic_pipeline.pk1")
```

```python
loaded_pipeline=joblib.load("titanic_pipeline.pk1")
```

```python
new_passenger=pd.DataFrame({
    'Pclass':[2],
    'Sex':['female'],
    'Age':[25],
    'Fare':[134.6],
    'Embarked':['S']
})
```

```python
prediction=loaded_pipeline.predict(new_passenger)
if prediction==0:
    print("Did not survive")
else:
    print("Survived")
```
