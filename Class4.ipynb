```python
numeric=df.iloc[:, 0:4].values
textual=df["species"]
```

```python
print(numeric)
```

```python
print(textual)
```

```python
means=np.mean(numeric,axis=0)
stds=np.std(numeric,axis=0)
#axis=0 means that we want to select from row 0, or from the start
print(means)
print(stds)
#feature with the highest std value is the most imp feature of the dataset
#this feature makes it easier to make predictions
#feature with low std makes it harder to make predictions
#we can see that the petal length is the most imp feature here at it has the hightest std value
```

```python
max=np.max(numeric,axis=0)
min=np.min(numeric,axis=0)
#min and max values helps find the range of a feature and also helps find outliers
#outliers are the impossible values of a feature, i.e -ve height, etc.
print(max)
print(min)
```

```python
#filteringn helps us find the outliers. if there are only a few values that are close to the min or max value, then they are the outliers
sepal_long=numeric[numeric[: ,2]>4]
#in the above cell, we see that the max sepal lenght is 4.4 so we set a filter at 4 to see how many entries have sepal length close to the max value
print(sepal_long)
#since there are a lot of value, we say that it is not an outlier
```

```python
#checking for petal length
petal_long=numeric[numeric[: ,2]>6]
print(petal_long)
```

```python
#now that we know that petal length is an imp feature
#and that it is not an outlier
#we will take the mean and std of each specie individually
for species in np.unique(textual):
    group=numeric[textual==species]
    print(species)
    print("Mean petal length:", group[:,2].mean())
    print("Std of petal length:", group[:,2].std())

#this helps us determine through our human intelligence that out of the three species, the one with the
#longest petal is virginica, the one with the smallest petals is setosa and versicolor lies in the middle
#this information helps us train models in identifying flowers by looking at the petal length as well
```
