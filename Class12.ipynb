```python
import numpy as np 
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.dummy import DummyClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix,classification_report
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
```

```python
titanic=pd.read_csv("/kaggle/input/datasets/fatuzahra/titanic/titanic.csv")
titanic.head()
```

```python
titanic["Age"].fillna(
    titanic["Age"].mean(),inplace=True
)
titanic.isnull().sum()
```

```python
titanic["Embarked"].fillna(
    titanic["Embarked"].mode()[0], inplace=True
)
titanic.isnull().sum()
```

```python
titanic= titanic.drop(columns="Cabin")
```

```python
titanic.isnull().sum()
```

```python
titanic=titanic[['Survived','Pclass','Age','Fare','Sex']]

titanic['Age'].fillna(titanic['Age'].median(),inplace=True)
```

```python
titanic['Sex']=titanic['Sex'].map({'male':0,'female':1})
```

```python
X= titanic.drop('Survived',axis=1)
Y=titanic['Survived']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y,
    test_size=0.2,
    random_state=42
)
```

```python
models={
    "KNN": KNeighborsClassifier(n_neighbors=5),
    "Decision Tree": DecisionTreeClassifier(max_depth=5),
    "Random Forest": RandomForestClassifier(n_estimators=100)
}
```

```python
for name, model in models.items():
    model.fit(X_train,Y_train)
    predics=model.predict(X_test)
    accuracy=accuracy_score(Y_test,predics)
    print(f'{name} Accuracy:',accuracy)
```

```python
scaler=StandardScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)

knn_scaled=KNeighborsClassifier(n_neighbors=5)
knn_scaled.fit(X_train_scaled,Y_train)

predics_knn_scaled=knn_scaled.predict(X_test_scaled)
accuracy_knn_scaled=accuracy_score(Y_test,predics_knn_scaled)

print("KNN Accuracy after scaling:",accuracy_knn_scaled)
```

```python
# Hyper Parameter Tuning:

# Settings/configuration of an algorithm that is set before training


# In logistic regression, weights are used that are automatically set by the model during training

# Decision Trees and Random Forest have hyperparameters that are set/defined/chosen by developer
```

```python
# Perfect balance of settings of a model ensures maximum optimization of the model.
# It helps prevent model/data memorization (overfitting)


# There are two techniques of Hyper Parameter Tuning

# 1. GridSearchCV
# 2. RandomizedSearchCV
```

```python
# GridSearchCV:

# It is suitable for small datasets.
# It is good because it chooses the best parameters settings values from the options we give it.
# Works automatically

# Trains and Tests model for all the possible parameters combination 
# Chooses the best settings/combination on the basis of score we provide it, in this case accuracy
# All this is done automatically
```

```python
param_grid={
    'n_estimators':[50,100],
    'max_depth':[None,5,10],
    'min_samples_split':[2,5]
}

rf=RandomForestClassifier(random_state=42)
```

```python
grid=GridSearchCV(
    rf,
    param_grid,
    cv=5,
    scoring="accuracy"
)
```

```python
# Scoring specifies what quality the model will be compared by.

# CV means cross validation
# it is mostly set as 5.

# As we know in train-test split, in some edge cases, data is divided unevenly
# The test data is too easy while the training data is too complex
# This can result in incorrect/faulty training

# To solve this problem we use CV as a model performance evaluation technique

# CV:
# Consists of multiple data splits
# cv=5 means data will ve divided into 5 parts
# Then one part of the data will be reserved for testing while the rest of the four will be used for training
# This part is then tested and test accuracy is calculated for it
# This process is repeated until each of the 5 parts have been tested for the rest of the 4
# Then the avg of these test accuracies is calculated
# This avg is more reliable than the train-test split accuracy
```

```python
grid.fit(X_train,Y_train)
```

```python
print('Best parameters:',grid.best_params_)
print('Best Accuracy:',grid.best_score_)
```

```python
# Drawback:

# 1.GridSearchCV is good but it is very slow for large datasets 
# because it trains and tests model for each possible combination of paramteres

# 2.Uses too many resources
```

```python
# so RandomizedSearchCV is preffered

# RandomizedSearchCV:
# Used for large datasets
# It does not train model for all possible datasets but instead randomly chooses a bunch of combinations
```
