```python
# iloc means integer based indexing

# loc means label based indexing

# iloc is used to select data by their number of rows and columns
# for example:

titanic.iloc[:5, :3]
# selects first five rows and first 3 columns. the last indices (5, 3) are exclusive
```

```python
# example of loc:

#display() is a built-in method to display result

display(titanic.loc[titanic["Pclass"]==1,["Name","Sex"]].head())

# returns the name and the sex of the first five passengers with Pclass 1
```

```python
# we can also concatnate two features together. this helps us decrease the size of columns in a dataset
# it is preferably done when two features are unnecessarily divided
# for example in out dataset, Sibsp and Parch are two different columns even tho they can be written as one Family
# so we concatenate them 

titanic["Family"]=titanic["SibSp"] + titanic["Parch"]+1
display(titanic["Family"].head())
```

```python
# we can also drop unnecessary columns from a dataset that are of no use:
# combining and dropping unnecessary features is known as feature engineering

titanic.drop(columns="Family",inplace=True)
#inplace=True is written so that changes are made on spot
display(titanic.columns)
```
