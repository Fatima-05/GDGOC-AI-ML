```python
score=[23,12,10,3]
print(np.mean(score))
print(np.median(score))
print(np.std(score))
print(np.min(score))
print(np.max(score))
```

```python
#std is the measure of how far each vlue is from the mean value
#std is used to check which data is better for computing
#it helps determine which data to eliminate and how and what data to clean
#std value also tells us how far or close the values are from each other
```

```python
#for example, to data can have the same mean/avg value but their std value helps us determine which data is cleaner and more suitable
score1=[60,70,80]
score2=[30,70,110]
#score1 and score2 both have the same avg but their std value is very different
print("mean and std of score1:")
print(np.mean(score1))
print(np.std(score1))
print("mean and std of score2:")
print(np.mean(score2))
print(np.std(score2))
```

```python
temps=np.array([30,31,29,32,28,32,33])
print(temps.mean())
print(temps.std())
#now since the std value is small, we can say that the values are close to each other
#this helps us come to the conclusion that the weather remained pretty consistent throughout the week
```
