```python
import numpy as np 
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.dummy import DummyClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix,classification_report
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
titanic=pd.read_csv("/kaggle/input/datasets/fatuzahra/titanic/titanic.csv")
titanic.head()
```

```python
titanic["Age"].fillna(
    titanic["Age"].mean(),inplace=True
)
titanic.isnull().sum()
```

```python
titanic["Embarked"].fillna(
    titanic["Embarked"].mode()[0], inplace=True
)
titanic.isnull().sum()
```

```python
titanic= titanic.drop(columns="Cabin")
```

```python
titanic.isnull().sum()
```

```python
titanic=titanic[['Survived','Pclass','Age','Fare','Sex']]

titanic['Age'].fillna(titanic['Age'].median(),inplace=True)
```

```python
titanic['Sex']=titanic['Sex'].map({'male':0,'female':1})
```

```python
X= titanic.drop('Survived',axis=1)
Y=titanic['Survived']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y,
    test_size=0.2,
    random_state=42
)
```

```python
# KNN (KNearestNeighbors)
# Used in classification as well as regression problems
# Model does not learn during training, it learns/make decision during prediction time
# This is why it is known as "Lazy Algorithm"
# Good results for small datasets
# Becomes slow when working with large datasets

# During predictiosn, it calculates the data point's distance to every other data point.
# Then, it selects k number of data points that are neighboring it (nearest to it)
# Then it predicts on the basis of majority label class among those neighbors.

# The value of k is specified manually. 
# It should neither be too small (leads to overfitting)
# Nor too large (leads to underfitting)
# ideal value is between 5-8, preferably 5.

# It uses the distance formula: sqrt((x2-x1)^2+(y2-y1)^2)

# For KNN, scaling is a crucial step because it predicts on the basis of distance
# If the ranges vary, the model becomes heavily biased as distance increases with range
# Without scaling, KNN often fails. 
# Below we will see how scaling improved the accuracy of KNN
```

```python
# Decision Trees:

# It works like a flowchart
# Filters out features on the basis of a flowchart
# Scaling is not necessary.
# Depth is specified for complexity control. 5 is a prefered value.

# The feature with the highest information gain is set on the top of the tree, and then so on.
# For information gain, first the Information Entropy is calculated
# Then it is used to calculate information gain
```

```python
# Random Forest:

# Advanced version of decision tree
# Consists of multiple trees
# Suitable for large datasets

# Multiple decision trees are evaluated, the majority output is set as the final prediction
# Prediction on the basis of majority voting
# i.e. if the result of two DT's is 0 and result of one DT is 1. the final prediction will be 0

# Pros:
# Random Forest is better because it checks different possibilities before predicting.
# This makes it more reliable
# It's accuracy is higher
# Less overfitting problem
# Scaling is not necessary
# It is fast

# Cons:
# Consumes large amount of memory
```

```python
models={
    "KNN": KNeighborsClassifier(n_neighbors=5),
    "Decision Tree": DecisionTreeClassifier(max_depth=5),
    "Random Forest": RandomForestClassifier(n_estimators=100)
}
```

```python
for name, model in models.items():
    model.fit(X_train,Y_train)
    predics=model.predict(X_test)
    accuracy=accuracy_score(Y_test,predics)
    print(f'{name} Accuracy:',accuracy)
```

```python
scaler=StandardScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)

knn_scaled=KNeighborsClassifier(n_neighbors=5)
knn_scaled.fit(X_train_scaled,Y_train)

predics_knn_scaled=knn_scaled.predict(X_test_scaled)
accuracy_knn_scaled=accuracy_score(Y_test,predics_knn_scaled)

print("KNN Accuracy after scaling:",accuracy_knn_scaled)
```
