```python
import numpy as np 
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.dummy import DummyClassifier
from sklearn.linear_model import LogisticRegression
```

```python
titanic=pd.read_csv("/kaggle/input/datasets/fatuzahra/titanic/titanic.csv")
titanic.head()
```

```python
cabin_dropped= titanic.drop(columns="Cabin")
```

```python
cabin_dropped["Age"].fillna(
    cabin_dropped["Age"].mean(),inplace=True
)
```

```python
cabin_dropped["Embarked"].fillna(
    cabin_dropped["Embarked"].mode()[0], inplace=True
)
cabin_dropped.isnull().sum()
```

```python
Q1=cabin_dropped["Fare"].quantile(0.25) #as it's the 25th percentile
Q3=cabin_dropped["Fare"].quantile(0.75) #as it's the 75th percentile

print(Q1)
print(Q3)

IQR=Q3-Q1

print(IQR)

lower_bound=Q1-1.5*IQR
upper_bound=Q3+1.5*IQR

print(lower_bound)
print(upper_bound)

titanic_clean=cabin_dropped[
(cabin_dropped["Fare"]>=lower_bound)&(cabin_dropped["Fare"]<=upper_bound)
]

#now we can visibly see that the outlier data was dropped

titanic_clean.shape
```

```python
titanic=titanic[['Survived','Pclass','Age','Fare','Sex']]

titanic['Age'].fillna(titanic['Age'].median(),inplace=True)
```

```python
titanic['Sex']=titanic['Sex'].map({'male':0,'female':1})
```

```python
X= titanic.drop('Survived',axis=1)
Y=titanic['Survived']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y,
    test_size=0.2,
    random_state=42
)
```

```python
baseline_model= DummyClassifier(strategy='most_frequent')
baseline_model.fit(X_train,Y_train)
```

```python
model=LogisticRegression(max_iter=1000)
```

```python
model.fit(X_train,Y_train)
```

```python
print("Weights:",model.coef_)
print("Bias",model.intercept_)
```

```python
# Predictions

y_pred=model.predict(X_test)

y_prob=model.predict_proba(X_test)
```

```python
print("Predictions",y_pred[:5])
print("Probabilities:",y_prob[:5])
```

```python
# Classification metrics:
# 1. Accuracy
# 2. Confusion Matrix
# 3. Precision
# 4. Recall
# 5. F1 Score
# 6. ROC Curve
# 7. Log Loss
```

```python
# Regression metrics:
# 1. Mean Absolute Error
# 2. Mean Squared Error
# 3. Root Mean Squared Error
# 4. R^2 Score
# 5. Adjusted R^2 Score
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix,classification_report
```

```python
# Accuracy is the measure of how correct the model is overall

# Accuracy = correct predictions/total no. of predictions
```

```python
accuracy=accuracy_score(Y_test,y_pred)
print("Accuracy:",accuracy)
```

```python
# Y_test contains actual values. By comparing actual values to predicted values, 
# We can find the %age of accuracy/coorect predictions
```

```python
# Accuracy can be misleading sometimes, especially when dataset is imbalanced
# So accuracy alone is not enough to evalute the model
# We consider other metrics as well
# For Example Cconfusion Matrix, Classification Report.
```

```python
# Confusion Matrix
# Gives information about:
# 1. True Positives: actual value is 1, predicted value is 1
# 2. True Negatives: actuall value is 0, predicted value is 0
# 3. False Positives: actual value is 0, predicted value is 1
# 4. False Negatives: actual value is 1, predicted value is 0

# Tips:
# If FN is high, threshold should be adjusted
# If FP is high, a different model/algorithm should be used
```

```python
cm=confusion_matrix(Y_test,y_pred)
print("Confusion matrix:",cm)
```

```python
# Classification Report
# Give information about:
# 1. Precision: how reliable the model is. what %age of each feature was predicted correctly
# 2. Recall: how good the actual detection of the model is
# 3. F1 Score: a balanced result b/w precision and recall. Harmonic mean b/w precision and recall
# 4. Support: tells us about the individual rows

# Tips:
# If accuracy and recall both are low, a different mode/algorithm should be used
# If accuracy is good but recall is bad, threshold should be adjusted or diff model/algo can be used
```

```python
c_report=classification_report(Y_test,y_pred)
print("Classification report:",c_report)
```

```python
predictions=baseline_model.predict(X_test)
baseline_accuracy=accuracy_score(Y_test,predictions)
print("Baseline Model Accuracy:",baseline_accuracy)
```

```python
# Tips:
# If accuracy is near baseline accuracy, then features should be worked on, feature engineering should be performed
# If accuracy is lower than baseline accuracy, the model should not be considered at all
# If accuracy is higher than baseline accuracy, the model can be considered for use 
# But other models' accuracy should be checked as well to find the best accuracy
```

```python
y_train_pred=model.predict(X_train)
test_accuracy=accuracy_score(Y_train,y_train_pred)
print("Test accuracy:",test_accuracy)
```
