```python
import numpy as np 
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic=pd.read_csv("/kaggle/input/datasets/fatuzahra/titanic/titanic.csv")
titanic.head()
```

```python
cabin_dropped= titanic.drop(columns="Cabin")
```

```python
cabin_dropped["Age"].fillna(
    cabin_dropped["Age"].mean(),inplace=True
)
```

```python
cabin_dropped["Embarked"].fillna(
    cabin_dropped["Embarked"].mode()[0], inplace=True
)
cabin_dropped.isnull().sum()
```

```python
Q1=cabin_dropped["Fare"].quantile(0.25) #as it's the 25th percentile
Q3=cabin_dropped["Fare"].quantile(0.75) #as it's the 75th percentile

print(Q1)
print(Q3)

IQR=Q3-Q1

print(IQR)

lower_bound=Q1-1.5*IQR
upper_bound=Q3+1.5*IQR

print(lower_bound)
print(upper_bound)

titanic_clean=cabin_dropped[
(cabin_dropped["Fare"]>=lower_bound)&(cabin_dropped["Fare"]<=upper_bound)
]

#now we can visibly see that the outlier data was dropped

titanic_clean.shape
```

```python
# if the distribution of data is not proper it means that the data is skewed
# it is either left-skewed (more data on the left)
# or right skewed (more data on the right)
# feature selection
# encoding
# train-test split
# baseline model

#  SkiKit learn 

# training - algos
# preprocessign - scaling, feature extration, feature selection, ecoding
# evaluation - R2 score, accuracy
# model selection - train-test split, R2 score
# optimization
```

```python
# importing SkiKit lean

from sklearn.model_selection import train_test_split
from sklearn.dummy import DummyClassifier
# feature selection

# to avoid unnecessary/unimportant/meaningless features
# choose features on the basis of following
# - relevant
# -interpretable (that can be evaluated or can be understandable by the model)

# methods to filter

# 1. correlation - heatmap (direct and inverse)
# 2. wrapper methods/ embedded methods:
#     different combos of features are tested


titanic=titanic[['Survived','Pclass','Age','Fare','Sex']]

titanic['Age'].fillna(titanic['Age'].median(),inplace=True)
```

```python
# encoding

# categorical data

# types

# norminal - doesn't need to be in a specific order i.e city name, sex
# ordinal - needs to be in order i.e education level

# - Label encoding
# - One-Hot encoding
# - Ordinal encoding
# - Target encoding
# - Binary encoding
# - Hash encoding


# Nominal, few categories -> best approach-> One-Hot
# Nominal, many categories -> best approach -> Target, Binary, Hash
# Ordinal -> best approach -> Label, Ordinal

titanic['Sex']=titanic['Sex'].map({'male':0,'female':1})
```

```python
X= titanic.drop('Survived',axis=1)
Y=titanic['Survived']

# X are the features and Y are the labels
```

```python
# train-test split

# generalization - trainng accuracy and unseen data test accuracy both are same
# overfitting - memorizes the dataset, training accuracy is very good, unseen data test accuracy is poor 
# underfitting - training accuracy and unseen dataset accuracy both are poor 

X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y,
    test_size=0.2,
    random_state=42
)
```

```python
# baseline model - benchmark, standard

baseline_model= DummyClassifier(strategy='most_frequent')
baseline_model.fit(X_train,Y_train)
```

```python
# SUpervised learning -> uses both features and labels to make predictions
# Supervised learning types:

# 1. Classification - prediction in form of category. yes/no. whether something happens or not i.e whether a person survives or not
# 2. Regression - predictions in form of continuous values. i.e temperature, marks prediction
```

```python
# Logistic regression
# Easy
# Commonly used
# used for Binary classificiation problems
# output in form of probability
# simple
```

```python
# Training through logistic regression means that a value i calculated against each row
```

```python
# weigth: it is the measure of importance of a feature

# score= (feature1*weight1)+(feature2*weight2)+(featureN*weightN)+bias(intercept)

# z= w1x1 + w2x2 + wNxN + b

# bias intercept is the starting point
```

```python
# sigmoid function is used to calculate probability

# y= 1/1+e^-(a0+a1x)
```

```python
# Probability is calculated and splitted into two categories, either yes or no. 0 or 1. either something will happen or not

# The point where this probability is divided is called Threshold Point
# The values above this point are grouped into one category
# The values below this point are grouped into the other category
# By default the value of threshold point is 0.5 (as probability can either be 0 or 1)
```

```python
# During training, th model takes bunch of random values to make predictions
# It then calculates error for each of these values

# error: difference between the actual value and the predicted value

# Then the weights of the features are calculated according to these errors

# This process is iterated a number of times until errors are minimum. This is known as training
                        
# The more it is iterated, the more accurate results are yeilded
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
# Model Initialization

model=LogisticRegression(max_iter=1000)

# max_iter is used ot speficify the number of times the process to calculate weights will be iterated
```

```python
# Model Training

model.fit(X_train,Y_train)

#only the training values are passed as the model is being trained right now
```

```python
print("Weights:",model.coef_)
print("Bias",model.intercept_)

# The -ve value show the inverse relation and the positive value shows the direct relation between a featur and the prediction
```

```python
# Predictions

y_pred=model.predict(X_test)

# During training we only pass the training features. labels are not passed

# Calculating probability

y_prob=model.predict_proba(X_test)
```

```python
# Using slicing to only see the first five predictions and probabilities

print("Predictions",y_pred[:5])
print("Probabilities:",y_prob[:5])

# The result shows us that the model predicted that the first three did not survive, then 4th and 5th survived

# Probability shows us that the first entity had a 0.89 probability of not surviving and only 0.10 probability of surviving, hence the prediction that they did not survive
```
