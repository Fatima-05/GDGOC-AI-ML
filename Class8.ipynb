```python
import numpy as np 
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic=pd.read_csv("/kaggle/input/datasets/fatuzahra/titanic/titanic.csv")
titanic.head()
```

```python
cabin_dropped= titanic.drop(columns="Cabin")
```

```python
cabin_dropped["Age"].fillna(
    cabin_dropped["Age"].mean(),inplace=True
)
```

```python
cabin_dropped["Embarked"].fillna(
    cabin_dropped["Embarked"].mode()[0], inplace=True
)
cabin_dropped.isnull().sum()
```

```python
Q1=cabin_dropped["Fare"].quantile(0.25) #as it's the 25th percentile
Q3=cabin_dropped["Fare"].quantile(0.75) #as it's the 75th percentile

print(Q1)
print(Q3)

IQR=Q3-Q1

print(IQR)

lower_bound=Q1-1.5*IQR
upper_bound=Q3+1.5*IQR

print(lower_bound)
print(upper_bound)

titanic_clean=cabin_dropped[
(cabin_dropped["Fare"]>=lower_bound)&(cabin_dropped["Fare"]<=upper_bound)
]

#now we can visibly see that the outlier data was dropped

titanic_clean.shape
```

```python
# if the distribution of data is not proper it means that the data is skewed
# it is either left-skewed (more data on the left)
# or right skewed (more data on the right)
# feature selection
# encoding
# train-test split
# baseline model

#  SkiKit learn 

# training - algos
# preprocessign - scaling, feature extration, feature selection, ecoding
# evaluation - R2 score, accuracy
# model selection - train-test split, R2 score
# optimization
```

```python
# importing SkiKit lean

from sklearn.model_selection import train_test_split
from sklearn.dummy import DummyClassifier
# feature selection

# to avoid unnecessary/unimportant/meaningless features
# choose features on the basis of following
# - relevant
# -interpretable (that can be evaluated or can be understandable by the model)

# methods to filter

# 1. correlation - heatmap (direct and inverse)
# 2. wrapper methods/ embedded methods:
#     different combos of features are tested


titanic=titanic[['Survived','Pclass','Age','Fare','Sex']]

titanic['Age'].fillna(titanic['Age'].median(),inplace=True)
```

```python
# encoding

# categorical data

# types

# norminal - doesn't need to be in a specific order i.e city name, sex
# ordinal - needs to be in order i.e education level

# - Label encoding
# - One-Hot encoding
# - Ordinal encoding
# - Target encoding
# - Binary encoding
# - Hash encoding


# Nominal, few categories -> best approach-> One-Hot
# Nominal, many categories -> best approach -> Target, Binary, Hash
# Ordinal -> best approach -> Label, Ordinal

titanic['Sex']=titanic['Sex'].map({'male':0,'female':1})
```

```python
X= titanic.drop('Survived',axis=1)
Y=titanic['Survived']
```

```python
# train-test split

# generalization - trainng accuracy and unseen data test accuracy both are same
# overfitting - memorizes the dataset, training accuracy is very good, unseen data test accuracy is poor 
# underfitting - training accuracy and unseen dataset accuracy both are poor 

X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y,
    test_size=0.2,
    random_state=42
)
```

```python
# baseline model - benchmark, standard

baseline_model= DummyClassifier(strategy='most_frequent')
baseline_model.fit(X_train,Y_train)
```
